```python
print("Null Values:")
print(df.isnull().sum())  # Count missing values in each column
print("\nData Types:")
print(df.dtypes)  # Check data types of each column

print("\nSummary Statistics:")
print(df.describe())  # Summary statistics for numeric columns

print("\nDuplicate Rows:")
print(df.duplicated().sum())  # Count duplicate rows

if 'target_column' in df.columns:  # Replace with your target column name
    print("\nClass Distribution:")
    print(df['target_column'].value_counts())  # Class distribution for target variable



print("\nColumns in Dataset:")
print(df.columns)


print("\nCategorical Columns:")
print(df.select_dtypes(include=['object']).columns)  # List categorical columns

# Example of merging datasets:
df2 = pd.read_csv('data/credit_record.csv')
print("\nFirst Dataset Shape:", df.shape)
print("Second Dataset Shape:", df2.shape)

```
